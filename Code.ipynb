```python
# from datetime import datetime, date
import datetime
def from_dob_to_age(dob):
    if(pd.notnull(dob)):
      today = datetime.date.today()
      dob = datetime.datetime.strptime(dob, '%Y-%m-%d')
      return today.year - dob.year - ((today.month, dob.day) < (dob.month, dob.day))
    else:
      return 0
    # return today.year - dob.year
```

### Calculating & Appending Duration of Medications 

```python
def duration(date):
    date = date[:-10] 
    if(len(date)>9):
      date = datetime.datetime.strptime(date, '%Y-%m-%d')
      start_date = datetime.date(date.year, date.month, date.day)
      end_date = datetime.date.today()
      time_difference = end_date - start_date
      age = time_difference.days
      return age
    else:
      return 0
```

```python
r3['DURATION_1'] = r3['MED_START'].apply(lambda x:duration(x))
r3['DURATION_2'] = r3['MED_STOP'].apply(lambda x:duration(str(x)))
r3['DURATION'] = r3['DURATION_1']-r3['DURATION_2']
r3 = r3.drop(['DURATION_1','DURATION_2'],axis=1)
r3.head()
```
